```python
import pandas as pd
import os
import warnings
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import ast

warnings.filterwarnings("ignore")
plt.style.use('dark_background')
```

```python
# Use this chunk to generate the dataset if not already created
# df = pd.DataFrame()
# files = os.listdir("./data/patients_new")
# for file in files:
    # patient = pd.read_json(os.path.join("./data/patients_new/", file))
    # df = pd.concat([df, patient], ignore_index=True)
# df.reset_index(drop=True)
# df.to_csv("./collated_dataset.csv", index=False)

df = pd.read_csv("./collated_dataset.csv")
```

```python
# Normalize admit and discharge times, and calculate length of stay
df = df.sort_values(by=["admittime"]).reset_index(drop=True)
df['admittime_wy'] = df['admittime'].apply(lambda x: pd.to_datetime(x).replace(year=2024))
df['dischtime_wy'] = df['dischtime'].apply(lambda x: pd.to_datetime(x).replace(year=2024))
df['dischtime_wy'] = df.apply(lambda row: pd.to_datetime(row['dischtime_wy']).replace(year=2025) if row['dischtime_wy'].month < row['admittime_wy'].month else row['dischtime_wy'], axis=1)
df["length_of_stay"] = pd.to_datetime(df["dischtime_wy"]) - pd.to_datetime(df["admittime_wy"])
df.head()
```

```python
# Take the sub-datasets (patients, hosp, poe, etc.), flatten them and append them to the main dataframe
def normalize_jsons(column, dropper, sorter):
    data = []
    # for each row's json
    for datum in column:
        # convert the first json in the list to a set of df columns
        json_df = pd.json_normalize(ast.literal_eval(datum)[0])
        data.append(json_df)
    # compile to a df, drop duplicates and reset index
    column_df = pd.concat(data, ignore_index=True).drop_duplicates(subset=dropper).sort_values(by=sorter).reset_index(drop=True)
    return column_df
```

```python
df = pd.merge(df, normalize_jsons(df["patients"], ["subject_id"], ["anchor_year", "subject_id"]), how='left', on="subject_id", suffixes=("", ""))
df.head()
```

```python
def plot_histogram(x1, bins, title, x_label, y_label, color1, lcolor1, x2=None, color2=None, lcolor2=None):
    ax = x1.hist(bins=bins, xlabelsize=10, ylabelsize=6, color=color1)
    plt.axvline(x1.median(), color=lcolor1, linestyle='dashed', linewidth=2)

    if x2 is not None:
        ax2 = x2.hist(bins=bins, xlabelsize=10, ylabelsize=6, color=color2)
        plt.axvline(x2.median(), color=lcolor2, linestyle='dashed', linewidth=2)

    ax.set_title(title, weight='bold')
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    plt.grid(False)
    plt.show()
```

```python
# Visualize admit and discharge time deltas (length of stay) for all patients. Length of stay is a good indicator of the value of care derived by class of patient.
plt.scatter(pd.to_datetime(df['dischtime_wy']), df['admittime_wy'], color="firebrick", s=12)
plt.scatter(pd.to_datetime(df['admittime_wy']), df['admittime_wy'], color="steelblue", s=12)

plt.xlabel('Admit & Discharge Time')
plt.ylabel('Admit Time')
plt.title('Admission and Discharge Times', weight="bold")
plt.xticks(rotation=45)
plt.show()
```

```python
# We'll explore length of stay based on four demographic indicators: class, race, gender and age.
# Race and gender are provided in the data. We'll use insurance (Medicaid or not) as a proxy for class. Age has been obfuscated in the dataset to "anchor age", which is approximately accurate.
```

```python
# We'll look at the distribution of ages to make sure it's about correct.
# Note that the spike at age 91 is because everyone above 89 is assigned 91 for privacy.
plot_histogram(df["anchor_age"], len(df["anchor_age"].value_counts()), "Distribution of ages across stays and patients", "Anchor age", "Frequency", "steelblue", "aqua", df.drop_duplicates(subset="subject_id")["anchor_age"], "firebrick", "red")
```

```python
# Medicaid patients are discharged, on average, slightly quicker than other patients
medicaid_df = df[df["insurance"] == "Medicaid"]
main_df = df[~df.index.isin(medicaid_df.index)]

plot_histogram((df["length_of_stay"].dt.total_seconds() / 86400), 12, "Length of stay of all patients vs Medicaid insured patients", "Duration (days)", "Frequency", "steelblue", "aqua", (medicaid_df["length_of_stay"].dt.total_seconds() / 86400), "firebrick", "red")

print(f"Non-Medicaid-insured patients: {main_df['length_of_stay'].mean()}")
print(f"Medicaid-insured patients: {medicaid_df['length_of_stay'].mean()}")
```

```python
# Non-white patients are discharged, on average, significantly quicker than white patients
non_white_df = df[~df["race"].str.contains("WHITE")]
main_df = df[~df.index.isin(non_white_df.index)]

plot_histogram((df["length_of_stay"].dt.total_seconds() / 86400), 12, "Length of stay of all patients vs non-white patients", "Duration (days)", "Frequency", "steelblue", "aqua", (non_white_df["length_of_stay"].dt.total_seconds() / 86400), "firebrick", "red")

print(f"White patients: {main_df['length_of_stay'].mean()}")
print(f"Non-white patients: {non_white_df['length_of_stay'].mean()}")
```

```python
# Female patients are discharged much quicker than male patients
female_df = df[df["gender"] == "F"]
main_df = df[~df.index.isin(female_df.index)]

plot_histogram((df["length_of_stay"].dt.total_seconds() / 86400), 12, "Length of stay of female patients vs male", "Duration (days)", "Frequency", "steelblue", "aqua", (female_df["length_of_stay"].dt.total_seconds() / 86400), "firebrick", "red")

print(f"Male patients: {main_df['length_of_stay'].mean()}")
print(f"Female patients: {female_df['length_of_stay'].mean()}")
```

```python
# There doesn't appear to be a significant trend of duration with age past 45 or so
old_df = df[df["anchor_age"] >= 65]
main_df = df[~df.index.isin(old_df.index)]

plt.scatter(df["anchor_age"], (df["length_of_stay"].dt.total_seconds() / 86400), c=df["anchor_age"].apply(lambda x: "firebrick" if x >= 65 else "steelblue"))
plt.xlabel('Anchor Age')
plt.ylabel('Length of stay (days)')
plt.title('Length of stay vs Anchor Age', weight='bold')
plt.show()

print(f"Young patients: {main_df['length_of_stay'].mean()}")
print(f"Old patients: {old_df['length_of_stay'].mean()}")
```

```python
# We can now look at variation in stay duration based on other information, such as admission circumstances.
```

```python
# Patients admitted under emergency conditions are in hospital for significantly longer than other patients
emergency_df = df[(df["admission_type"].str.contains("EMER")) | (df["admission_type"].str.contains("URGENT"))]
main_df = df[~df.index.isin(emergency_df.index)]

plot_histogram((df["length_of_stay"].dt.total_seconds() / 86400), 12, "Length of stay of all patients vs emergency-admitted patients", "Duration (days)", "Frequency", "steelblue", "aqua", (emergency_df["length_of_stay"].dt.total_seconds() / 86400), "firebrick", "red")

print(f"Otherwise-admitted patients: {main_df['length_of_stay'].mean()}")
print(f"Emergency-admitted patients: {emergency_df['length_of_stay'].mean()}")
```

```python
# There is seasonality to stay duration: patients admitted in the summer months are discharged much quicker
summer_df = df[(df["admittime_wy"].dt.month >= 5) & (df['admittime_wy'].dt.month <= 9)]
main_df = df[~df.index.isin(summer_df.index)]

plot_histogram((df["length_of_stay"].dt.total_seconds() / 86400), 12, "Length of stay of summer admits vs other patients", "Duration (days)", "Frequency", "steelblue", "aqua", (summer_df["length_of_stay"].dt.total_seconds() / 86400), "firebrick", "red")

print(f"Patients admitted outside summer: {main_df['length_of_stay'].mean()}")
print(f"Patients admitted during summer: {summer_df['length_of_stay'].mean()}")
```

```python
# This could be due to many diseases such as CVD and respiratory illnesses hitting harder during the winter. If so, the elderly would be disproportionately affected. This doesn't seem to be the case, pointing to something else.
summer_df = df[(df["admittime_wy"].dt.month >= 5) & (df['admittime_wy'].dt.month <= 9)]
main_df = df[~df.index.isin(summer_df.index)]

plt.scatter(pd.to_datetime(df["admittime_wy"]), (df["length_of_stay"].dt.total_seconds() / 86400), c=df["anchor_age"].apply(lambda x: "firebrick" if x >= 65 else "steelblue"))
plt.xlabel('Admit Time')
plt.ylabel('Length of stay (days)')
plt.title('Length of stay against Admit Time & Age', weight='bold')
plt.xticks(rotation=45)
plt.show()

print(f"Mean age of non-summer admits: {main_df['anchor_age'].mean()}")
print(f"Mean age of summer admits: {summer_df['anchor_age'].mean()}")
```
